## Web Scraping

```python
#BeautifulSoup and Requests
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
#checks if server is available
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
# find and find all
```

```python
soup.find('div')
```

```python
#selecting a particular tag
soup.find_all('p',class_ = 'lead')
```

```python
#selecting a particular tag text
soup.find('p',class_ = 'lead').text.strip()
```

```python
soup.find_all('th')
```

```python
soup.find('th').text.strip()
```

```python
urlss= 'https://finance.yahoo.com/trending-tickers'
```

```python
pagess = requests.get(urlss)
```

```python
pagess
```

```python
soupsss= BeautifulSoup(pagess.text,'html')
```

```python
table_head = soupsss.find_all('th')
```

```python
table_head_fm = [ title.text.strip() for  title in table_head]
table_head_fm 
```

```python
print(table_head_fm )
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_head_fm )
df
```
