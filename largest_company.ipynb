```python
from bs4 import BeautifulSoup
import requests
```

```python
url= 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_='wikitable sortable sticky-header')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
 soup.find_all('th')[0:13]
```

```python
world_titles =  table.find_all('th')[0:9]
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
world_table_titles 
```

```python
world_table_titles[6] = 'Headquarters'
world_table_titles[3] = 'Revenue (USD millions)'
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    # Assuming you have extracted the data into individual_row_data
# Adjust the mapping to match the correct columns
    rank = individual_row_data[8]
    name = individual_row_data[1]
    industry = individual_row_data[2]
    revenue = individual_row_data[3]
    profit = individual_row_data[4]
    employees = individual_row_data[5]
    headquarters = individual_row_data[6]
    state_owned = individual_row_data[7]  # Assuming this should be here based on your example
    ref = individual_row_data[8]  # Assuming this should be here based on your example

# Create a new row in your DataFrame using the correct mapping
    length = len(df)
    df.loc[length] = [rank, name, industry, revenue, profit, employees, headquarters, state_owned, ref]

    #print((individual_row_data))
    length = len(df)
    df.loc[length] = individual_row_data
```
